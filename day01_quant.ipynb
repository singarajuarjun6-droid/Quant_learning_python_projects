```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
df = yf.download("AAPL", start="2020-01-01", end="2023-01-01")
```

```python
print(df.head)
```

```python
df['returns'] = df['Close'].pct_change()
```

```python
print(df)
```

```python
#calculating stats
mean = df['returns'].mean()
std = df['returns'].std()
```

```python
print("",mean,"\n",std)
```

```python
df["Close"].plot()
plt.title("AAPL Price Chart")
plt.savefig("aapl_price.png")
```

```python
df.to_csv("aapl_data_with_returns.csv")
```
